```python
import pandas as pd
import glob
```

```python
#-------------- read single file --------------

df0 = pd.read_csv('./data/JC-201601-citibike-tripdata.csv')
print(df0.head())    
print(df0.info())
```

```python
#-------------- concat all files --------------
csv_files = sorted(glob.glob('./data/*.csv'))[:11]

df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index=True)
df.to_csv('concat_data_citibike.csv', index=False)
df = pd.read_csv('concat_data_citibike.csv')

print(df.info())
# print(df.shape)
# print(df.describe())
# print(df.head(10)) 
```

```python
#--------------- Trip Duration -----------------
print(df['Trip Duration'].head(20))
print(df['Trip Duration'].dtype)
print(df['Trip Duration'].isnull().sum())
df['Trip Duration'] = pd.to_numeric(df['Trip Duration'], errors='coerce').fillna(0).astype(int)
```

```python
#----------------- Strat Time ----------------
print(df['Start Time'].head(20))
print(df['Start Time'].isnull().sum())
df['Start Time'] = df['Start Time'].dropna()
df['Start Time'] = pd.to_datetime(df['Start Time'])
print(df['Start Time'].dtype)
```

```python
#----------------- Stop Time ----------------
print(df['Stop Time'].head(20))
print(df['Stop Time'].isnull().sum())
df['Stop Time'] = pd.to_datetime(df['Stop Time'], errors='coerce')
df = df.dropna(subset=['Stop Time'])
df = df[df['Start Time'] <= df['Stop Time']]
```

```python
#----------------- Start Station ID ----------------
# print(df['Start Station ID'].head(20))
print(df['Start Station ID'].nunique())
print(df['Start Station ID'].unique())
print(df['Start Station ID'].isnull().sum())
df['Start Station ID'] = df['Start Station ID'].dropna()
df['Start Station ID'] = pd.to_numeric(df['Start Station ID'], errors='coerce').fillna(0).astype(int)
print(df['Start Station ID'].dtype)
```

```python
#----------------- Start Station Name ----------------
print(df['Start Station Name'].head(10))
print(df['Start Station Name'].nunique())
print(df['Start Station Name'].unique())
print(df['Start Station Name'].isnull().sum())
df['Start Station Name'] = df['Start Station Name'].dropna()
print(df['Start Station Name'].dtype)
```

```python
#-------------- Start Station Latitude ----------------
print(df['Start Station Latitude'].head(10))
print(df['Start Station Latitude'].count())
print(df['Start Station Latitude'].nunique())
print(df['Start Station Latitude'].unique())
print(df['Start Station Latitude'].isnull().sum())
df['Start Station Latitude'] = df['Start Station Latitude'].dropna()
df['Start Station Latitude'] = pd.to_numeric(df['Start Station Latitude'], errors='coerce').fillna(0.0)
```

```python
#------------------- Start Station Longitude -----------------
print(df['Start Station Longitude'].head(10))
print(df['Start Station Longitude'].count())
print(df['Start Station Longitude'].nunique())
print(df['Start Station Longitude'].unique())
print(df['Start Station Longitude'].isnull().sum())
df['Start Station Longitude'] = df['Start Station Longitude'].dropna()
df['Start Station Longitude'] = pd.to_numeric(df['Start Station Longitude'], errors='coerce').fillna(0.0)
```

```python
#----------------- End Station ID ----------------
print(df['End Station ID'].head(10))
print(df['End Station ID'].nunique())
print(df['End Station ID'].unique())
print(df['End Station ID'].isnull().sum())
df['End Station ID'] = df['End Station ID'].dropna()
df['End Station ID'] = pd.to_numeric(df['End Station ID'], errors='coerce').fillna(0).astype(int)
```

```python
#-------------------- End Station Name -------------------
print(df['End Station Name'].head(10))
print(df['End Station Name'].nunique())
print(df['End Station Name'].unique())
print(df['End Station Name'].isnull().sum())
df['End Station Name'] = df['End Station Name'].dropna()
```

```python
#--------------------- End Station Latitude ---------------------
print(df['End Station Latitude'].head(10))
print(df['End Station Latitude'].count())
print(df['End Station Latitude'].nunique())
print(df['End Station Latitude'].unique())
print(df['End Station Latitude'].isnull().sum())
df['End Station Latitude'] = df['End Station Latitude'].dropna()
df['End Station Latitude'] = pd.to_numeric(df['End Station Latitude'], errors='coerce').fillna(0.0)
```

```python
#------------------ End Station Longitude ---------------------
print(df['End Station Longitude'].head(10))
print(df['End Station Longitude'].count())
print(df['End Station Longitude'].nunique())
print(df['End Station Longitude'].unique())
print(df['End Station Longitude'].isnull().sum())
df['End Station Longitude'] = df['End Station Longitude'].dropna()
df['End Station Longitude'] = pd.to_numeric(df['End Station Longitude'], errors='coerce').fillna(0.0)
# print(df['End Station Longitude'].head(50))
```

```python
#---------------- Bike ID ---------------------
print(df['Bike ID'].head(10)) 
print(df['Bike ID'].count())
print(df['Bike ID'].nunique())
print(df['Bike ID'].unique())
print(df['Bike ID'].isnull().sum())
df['Bike ID'] = df['Bike ID'].dropna()
df['Bike ID'] = pd.to_numeric(df['Bike ID'], errors='coerce').fillna(0).astype(int)
```

```python
#-------------- User Type ---------------------
# print(df['User Type'].head(10))
print(df['User Type'].count())
print(df['User Type'].nunique())
print(df['User Type'].unique())
print(df['User Type'].isnull().sum())
df['User Type'] = df['User Type'].fillna('Unknown')
```

```python
#-------------- Birth Year ---------------------
print(df['Birth Year'].head(50))
print(df['Birth Year'].isnull().sum())
df['Birth Year'] = df['Birth Year'].replace('NaN', '0').fillna('0')
df['Birth Year'] = pd.to_numeric(df['Birth Year'], errors='coerce').fillna('0').astype(int)
print(df['Birth Year'].head(50))
```

```python
#--------------- Gender ---------------------
print(df['Gender'].head(10))
print(df['Gender'].nunique())
print(df['Gender'].unique())
df['Gender'] = pd.to_numeric(df['Gender'], errors='coerce').fillna(0).astype(int)

df['Gender'] = df['Gender'].replace({
    0: 'Unknown',
    1: 'Male',
    2: 'Female'
})

df['Gender'] = df['Gender'].astype(object)
print(df['Gender'].head(10))
```

```python
# data cleaning
df = df[df['Start Time'] <= df['Stop Time']]
df = df[df['Birth Year'] != 0]
df = df[df['Gender'] != 'Unknown']
df = df[df['User Type'] != 'Unknown']
df['Start Station Longitude'] = df['Start Station Longitude'].astype(float).abs()
df['End Station Longitude'] = df['End Station Longitude'].astype(float).abs()
```

```python
# print(df.info())
# print(df.head(50))
df2 = df.copy()  # keep cleaned data in df2

df2.to_csv('clean_data_citibike.csv', index=False)  # save to CSV

print(df2.info())
```
